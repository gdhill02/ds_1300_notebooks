```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
#Question 1 - Washington DC, New Jersey, Rhode Island, Puerto Rico, Massachusetts
facts['Pop_density_2020']=facts['Pop_2020']/facts['Area_land']
Q1=facts.sort_values(by=['Pop_density_2020'], ascending=False)
Q1.head(5)
```

```python
#Question 2 - American Samoa, Guam, N. Mariana Islands, US Virgin Islands
facts['Maritime']=facts['Area_water']>facts['Area_land']
facts[facts['Maritime']==True]
```

```python
#Question 3 - Texas, Florida, California, Georgia, Washington
facts['Growth']=facts['Pop_2020']-facts['Pop_2010']
Q3=facts.sort_values(by=['Growth'], ascending=False)
Q3.head(5)
```

```python
#Question 4 - Northern Mockingbird (population)
Q4=facts.groupby(by='State_bird').sum()
Q4.sort_values(by='Pop_2020', ascending=False).head(1)
```

```python
#Question 4 (cont.) - Willow Ptarmigan (land area)
Q4.sort_values(by='Area_land', ascending=False).head(1)
```

```python
#Question 5 - see below
facts[facts['Capital_pop']==facts['city_1_pop']]
```

```python
#Question 6 - New York
facts['range_in_city_pop']=100*(facts['city_1_pop']-facts['city_5_pop'])/(facts['city_1_pop'])
Q6=facts.sort_values(by=['range_in_city_pop'], ascending=False)
```

```python
dates=pd.read_csv('../data/state_dates.tsv',delimiter="\t")
```

```python
dates['Date_Fixed'] = pd.to_datetime(dates['Date'],errors='raise')
dates[['Date','Date_Fixed']]
```

```python
from datetime import datetime
```

```python
data=pd.merge(facts,dates,left_on='USPS_code',right_on='Abbreviation',how='outer')
```

```python
#Question 7 - Mountain Laurel
OG_13=data[data['Date_Fixed']<pd.datetime(1790,1,1)]
OG_13.groupby(by='State_flower').count()
```

```python
#Question 8 - Puerto Rico's density > Alaska's density 
data[data['Status'].str.contains('Territory')].sort_values(by='Pop_density_2020', ascending = False)
```

```python
data[data['Status'].str.contains('State')].sort_values(by='Pop_density_2020', ascending = True)
```
