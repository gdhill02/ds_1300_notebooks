```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

%config InlineBackend.figure_format ='retina'
```

```python
df = pd.read_csv('nba_95_96_pg-Copy1.csv',delimiter=',',engine='python')
```

```python
newset = df[df['MP']>15]
```

```python
newset.head(10)
```

```python
ax = newset.plot.scatter(x='Age', y="DRB", label='Defensive Rebounds', c='r', figsize=(20,10))
newset.plot.scatter(x='Age', y='ORB', label='Offensive Rebounds', figsize=(20,10), c='g', ax=ax, legend=True);
plt.savefig('.pdf', format = 'pdf')
```
